```python
import pandas as pd
```

```python
df = pd.read_csv('country_panel.csv')
```

```python
df.head()
```

```python
#locationの要素（国名）一覧をリストにする
locations = df['location'].unique().tolist()
```

```python
# location の国名を　location_idに変換する
#location_id列をlocation列をコピーして作る
df['location_id'] = df['location']
#locationの要素(国名)に出現順に番号を対応付ける辞書を作成する
ids = list(range(len(locations)))
dic = dict(zip(locations, ids))
dic
```

```python
#location_id列の国名を辞書を使ってidに変更する
df = df.replace({'location_id': dic})
```

```python
df.head()
```

```python
df.describe()
```

```python
#インデックスに、location_idとdateを文字列からdatetimeに変換したものを設定する
df = df.set_index(['location_id', pd.DatetimeIndex(df['date'])],drop=False)
```

```python
import statsmodels.api as sm
from linearmodels.panel import PooledOLS
```

```python
#説明変数を指定する
exog_vars = ["people_vaccinated", "total_boosters", "people_fully_vaccinated", "aged_70_older"]
exog = sm.add_constant(df[exog_vars])
#被説明変数には df.new_deaths を指定する
mod = PooledOLS(df.new_deaths, exog)
pooled_res = mod.fit()
print(pooled_res)
```

```python
#必要なパッケージをインポートする
from linearmodels.panel import PanelOLS

#説明変数を指定する
exog_vars = ["people_vaccinated", "total_boosters", "people_fully_vaccinated", "aged_70_older"]
exog = sm.add_constant(df[exog_vars])
#被説明変数には df.new_deaths を指定する
#Fixed Effect で分析するために、entity_effect=Trueを指定する
#mod = PanelOLS(df.new_deaths, exog, entity_effects=True)
#ではエラーが出るので、drop_absorbed=Trueを追加する。
mod = PanelOLS(df.new_deaths, exog, entity_effects=True, drop_absorbed=True)
pooled_res = mod.fit()
print(pooled_res)
```
